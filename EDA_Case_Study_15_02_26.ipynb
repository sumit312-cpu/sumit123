```python
# df["year"] -> mean
# df["Publisher"] -> mode
```

```python

# int -> mean
# obj -> mode

for col in df.columns:
  if(df[col].dtype == "object"):
    df[col] = df[col].fillna(df[col].mode()[0])
  else:
    df[col] = df[col].fillna(df[col].mean())
```
